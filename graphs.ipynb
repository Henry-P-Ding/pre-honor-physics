```python
def get_verts(start_vert, end_vert, num_vert):
    sample_verts_left = np.linspace(start_vert, end_vert, num_vert)
    sample_verts_right = sample_verts_left + (end_vert - start_vert) / (num_vert - 1)
    sample_verts = np.vstack([sample_verts_left, sample_verts_right])
    return sample_verts.T.flatten(), sample_verts_left, sample_verts_right
```

```python
t1 = np.linspace(0, 10, 51)
v1 = (np.sin(t1) + 2) * np.exp(t1 / 10)

fig, (ax1, ax2) = plt.subplots(1, 2)
ax1.plot(t1, v1, "black")
sample_verts, sample_verts_left, sample_verts_right = get_verts(0, 10, 11)
for verts_left in sample_verts_left:
    ax1.add_patch(Rectangle((verts_left, 0), 1, v1[5 * int(verts_left)], edgecolor="black"))

ax2.plot(t1, v1, "black")
sample_verts, sample_verts_left, sample_verts_right = get_verts(0, 10, 21)
for verts_left in sample_verts_left:
    ax2.add_patch(Rectangle((verts_left, 0), 0.5, v1[int(5* verts_left)], edgecolor="black"))

fig.set_size_inches(8, 3)
ax1.set_ylabel(r"$v$ (m/s)")
ax1.set_xlabel(r"$t$ (s)")
ax2.set_xlabel(r"$t$ (s)")
ax1.set_yticks(np.linspace(0, 8, 5))
ax1.set_yticks(np.linspace(0, 8, 5))
ax1.set_xticks(np.linspace(0, 10, 6))
ax2.set_yticks([])
ax2.set_xticks(np.linspace(0, 10, 6))
ax1.set_ylim(0, 8)
ax1.set_xlim(0, 10)
ax2.set_ylim(0, 8)
ax2.set_xlim(0, 10)
fig.tight_layout()
fig.tight_layout()
fig.savefig("./lecture/lecture2/figures/general-vt.pgf")
```
